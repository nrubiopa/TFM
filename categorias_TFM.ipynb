```python
import requests
import json
import math
import pandas as pd
import numpy as np

# Para este notebook solo
import scipy.stats as stats
```

```python
file_path = './clinicaltrials_limpio.csv'
df= pd.read_csv(file_path)
```

```python
categorical_variables = [
    "ArmGroupType", "Condition", "ConditionAncestorId", "ConditionAncestorTerm",
    "ConditionBrowseBranchAbbrev", "ConditionBrowseBranchName", "ConditionBrowseLeafAsFound",
    "ConditionBrowseLeafId", "ConditionBrowseLeafName", "ConditionBrowseLeafRelevance",
    "ConditionMeshId", "ConditionMeshTerm", "MinimumAge", "StdAge", "OrgClass",
    "HealthyVolunteers", "DesignInterventionModel", "DesignMasking", "DesignPrimaryPurpose",
    "InterventionType", "IsFDARegulatedDevice",
    "IsFDARegulatedDrug", "LeadSponsorClass", "LocationCountry", "HasExpandedAccess",
    "OverallOfficialRole", "OversightHasDMC", "Phase",
    "ResponsiblePartyType", "StudyType"
]



# Crear un diccionario para almacenar las tablas de frecuencia relativa
relative_frequency_tables = {}

# Calcular las tablas de frecuencia relativa para cada variable categ√≥rica
for var in categorical_variables:
    cross_tab = pd.crosstab(df[var], df['SuccessfulCases'], normalize='index') * 100
    relative_frequency_tables[var] = cross_tab

# Mostrando las tablas de frecuencia relativa
for var, table in relative_frequency_tables.items():
    print(f"Tabla de Frecuencia Relativa para '{var}':")
    display(table.style.background_gradient(cmap='coolwarm').format("{:.2f}%"))
```

```python
# Realizando la prueba de Chi-cuadrado para cada variable categ√≥rica
chi2_results = {}

for var in categorical_variables:
    contingency_table = pd.crosstab(df[var], df['SuccessfulCases'])
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    chi2_results[var] = {'Chi2 Value': chi2, 'P-Value': p, 'Degrees of Freedom': dof}

# Mostrando los resultados de las pruebas de Chi-cuadrado
chi2_results_df = pd.DataFrame(chi2_results).T
chi2_results_df
```
